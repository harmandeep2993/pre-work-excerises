```python
# Import dependencies
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Store the link of csv file
url= "https://raw.githubusercontent.com/data-bootcamp-v4/prework_data/main/students_performance.csv"

# Read the csv file from url and store it into dataframe "df"
df= pd.read_csv(url)

# Display top 5 rows of the dataset
df.head()
```

#### Numerical Variable Visualization

```python
# math score variable visualization
sns.histplot(data=df, x="math score")
```

**Obeservations**
- Most students scored between 50 and 80, showing that the majority achieved average to above-average results.
- Very few students scored below 30 or above 90, indicating fewer extreme low and high performers.

```python
# reading score	variable visualization
sns.histplot(data=df, x="reading score")
```

**Observations**
- Most students scored between 55 and 85, indicating generally strong reading performance.
- Very few students scored below 30 or above 95, showing limited extreme values.

```python
# writing score	variable visualization
sns.histplot(data=df, x="writing score")
```

**Observations**
- Most students scored between 55 and 85, indicating average to strong performance in writing.
- The highest concentration of scores appears around 65–75, which represents the central tendency.
- Very few students scored below 30 or above 95, showing limited extreme values.
