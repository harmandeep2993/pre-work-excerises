# List Exercise: Calculate Avergae 

Write a program that calculates the average of the elements in a given list. The list can contain both positive and negative numbers.

``Hint: you can either look for a Python function that calculates the average or look for a Python function that sums all elements in a list and divide by the length of the list.``

```python
import statistics as stats

# Define elements in the list
numbers = [2, 4, 6, 8, 10]

avg_of_elements =  stats.mean(numbers)

print(f"The average of the elements in the list: {avg_of_elements}")
```

```python
# Define variables
sum_of_elements=0
list_length= len(numbers)

# Loop through the elements of the list and add them
for i in range(list_length):
    sum_of_elements += numbers[i]

# Calculate the average
list_avg = sum_of_elements / list_length

print(f"The average of the elements in the list: {list_avg}")
```
