# List Data Sturcture Exercise

Given the list:

    lst = [1, 2, 34, 5, 3, 12, 9, 8, 67, 89, 98, 90, 39, 21, 45, 46, 23, 13]
    
Answer the following questions:
    
- How many elements are in the list?
- Using indexes, find out which is the first element in the list?
- Using indexes, find out which is the last element in the list?
- What is the index of element `90` in the list? 
- Which are the first 8 elements in the list?
- Append elements 100 and 110 to the list.
- Sort the elements in the list. 

```python
# Define the list
lst = [1, 2, 34, 5, 3, 12, 9, 8, 67, 89, 98, 90, 39, 21, 45, 46, 23, 13]
```

```python
# Length of the list
len(lst)
```

```python
# Frist element in the list
lst[0]
```

```python
# Last element in the list
lst[-1]
```

```python
# Index of element 90
pos= lst.index(90)

print(f"Position of 90 element in the list: {pos}")

# Pass the same position in list to verify; it should return 90
print(f"\nElement at {pos}th in the list:  {lst[pos]}")
```

```python
# First eight element in the list
lst[:8]
```

```python
# Append new element 100 & 110 in the list
lst.append(100)
lst.append(110)

print(lst)
```

```python
# Define new list of elements to add
number= [100, 110]

# Add new list to old list elements with + operator
lst= lst + number

# print elements in the list
print(lst)
```

```python
# Sort elements in the list
lst.sort()

print(lst)
```
