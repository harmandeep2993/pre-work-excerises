# List Exercise: Name Search

Write a program that searches for a specific element in a given list and returns its index. If the element is not found, print a message indicating that it is not present in the list.

``
names = ["Alice", "Bob", "Charlie", "Dave"] 
``

``
search_name = "Bob" # Element to search in "names" list
``


```python
# Define the list
names = ["Alice", "Bob", "Charlie", "Dave"]

search_name = "Bob"

if search_name in names:
    index_pos = names.index(search_name)
    print(f"The index position of {search_name} in the list is {index_pos}")
else:
    print(f"{search_name} is not in the list.")
```

## User Input and Handling case sentivity

```python
# Define the list
names = ["Alice", "Bob", "Charlie", "Dave"]

# Take user input
user_input= input("Enter name to search")

# Convert input to lowercase and store to new variable
search_name= user_input.lower()

# create lower version of the list
names_lower = [names.lower() for names in names]

if search_name in names_lower:
    index_pos= names_lower.index(search_name)
    print(f"The index position of {search_name} in the list is {index_pos}")

else:
    print(f"{search_name} is not in the list")
```
