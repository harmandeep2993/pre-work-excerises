## 4.3 Iterations Exercise on Dictionaries

```python
# Define given dictionary

word_freq = {'love': 25, 'conversation': 1, 'every': 6, "we're": 1, 'plate': 1, 'sour': 1, 
             'jukebox': 1, 'now': 11, 'taxi': 1, 'fast': 1, 'bag': 1, 'man': 1, 'push': 3,
             'baby': 14, 'going': 1, 'you': 16, "don't": 2, 'one': 1, 'mind': 2, 'backseat': 1,
             'friends': 1, 'then': 3, 'know': 2
}
```

**4.3.1** Iterate on the items of this dictionary to print only those keys where the frequency of the word is less than 3.

```python
# Create an empty list to store keys with frequency less than 3
dict_keys = []

# Loop through each key-value pair in the dictionary
for key, freq in word_freq.items():
    
    # Check if the frequency of the word is less than 3
    if freq < 3:
        
        # Add the key to the list if condition is true
        dict_keys.append(key)

# Print all keys whose frequency is less than 3
print(dict_keys)
```

**4.3.2** Iterate on the items of this dictionary to print the word with the highest frequency.

```python
# Initialize variables to track the highest frequency and corresponding word
max_freq = 0
max_word = ""

# Iterate through each key-value pair in the dictionary
for word, freq in word_freq.items():
    
    # Check if current frequency is greater than the maximum found so far
    if freq > max_freq:
        max_freq = freq
        max_word = word

# Print the word with the highest frequency
print("Word with highest frequency:", max_word)
print("Frequency:", max_freq)
```
