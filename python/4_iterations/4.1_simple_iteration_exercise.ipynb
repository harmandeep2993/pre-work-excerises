````python
# Loop through numbers from 10 to 30
for i in range(10, 31):
    
    # Check if the number is a multiple of 5
    if i % 5 == 0:
        print(i)
````

**4.1.5** Use the function input() to prompt the user to enter an integer number. Then calculate the factorial of that number using loops. To verify the correctness of your result, you can compare it with the factorial calculated using the factorial function from the math library.

````python
import math  # Import math library for verification

# Take integer input from the user
user_input = int(input("Enter a number to find factorial: "))

# Check for negative input
if user_input < 0:
    print("Factorial is not defined for negative numbers.")
else:
    fact = 1               
    temp = user_input      

    # Calculate factorial using a loop
    while user_input > 0:
        fact *= user_input
        user_input -= 1

    # Print factorial calculated using loop
    print("Factorial using loop:", fact)

    # Verify result using math.factorial()
    print("Factorial using math library:", math.factorial(temp))
````
