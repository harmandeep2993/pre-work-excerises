#### 4.2 List Iteration Exercise

```python
# Define given list

x = [12, 43, 4, 1, 6, 343, 10, 34, 12, 93, 783, 330, 896, 1, 55]
```

**4.2.1** Write code to find only those numbers in the list that are divisible by 3.

```python
# Create an empty list to store numbers divisible by 3
div_by_3 = []

# Loop through each number in the list x
for number in x:
    
    # Check if the number is divisible by 3
    if number % 3 == 0:
        div_by_3.append(number)

# Sort the final list once (outside the loop)
div_by_3.sort()

# Print the result
print(div_by_3)
```

**4.2.2** Write code to find only those numbers in a list that end with the digit 3. Create a new list with those numbers. To accomplish this, you can convert the integers to strings and use indexing to access the last digit. Here's an example of how to use indexing with strings:

```python
print(x)
```

```python
# Create an empty list to store numbers ending with digit 3
str_list = []

# Loop through each number in the list x
for num in x:
    
    # Convert the number to string and check if the last digit is '3'
    if str(num)[-1] == "3":
        
        # If condition is true, add the number to the new list
        str_list.append(num)

# Print the final list
print(str_list)
```

**4.2.3** Find the minimum value in the list x. You can research if there is any function available directly to calculate the minimum value.

```python
# Find the minimum value in the list x using the built-in function
min_value = min(x)

# Print the minimum value
print(min_value)
```

```python
# Assume the first element is the minimum initially
min_value = x[0]

# Loop through each number in the list
for num in x:
    
    # Update minimum if a smaller value is found
    if num < min_value:
        min_value = num

# Print the final minimum value
print(min_value)
```

**4.2.4** Find the maximum value in the list x. Similarly, research if there is any function available directly to calculate the maximum value

```python
# Find the maximum value in the list x using the built-in function
max_val = max(x)

# Print the maximum value
print(max_val)
```

```python
# Assume the first element is the maximum initially
max_value = x[0]

# Loop through each number in the list
for num in x:
    
    # Update maximum if a smaller value is found
    if num > max_value:
        max_value = num

# Print the final maximum value
print(max_value)
```
