# Grade Classification Exercise

- Write a program that prompts the user to enter a numerical grade and prints the corresponding grade classification according to the following criteria:

- Grades 90 and above: "A" Grades 80-89: "B" Grades 70-79: "C" Grades 60-69: "D" Grades below 60: "F"

```python
# Prompt the user to enter a numerical grade
user_input_grade = float(input("Enter your numerical grade: "))

# Determine the grade classification
if user_input_grade >= 90:
    print("Grade: A")

elif 80 <= user_input_grade < 90:
    print("Grade: B")

elif 70 <= user_input_grade < 80:
    print("Grade: C")

elif 60 <= user_input_grade < 70:
    print("Grade: D")

else:
    print("Grade: F")
```
