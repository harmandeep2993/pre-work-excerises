# Larger Number Exercise

1. Write a program that prompts the user to enter two numbers and determines which number is larger or if they are equal.

```python
# Take first input from the user
user_input_number_1 = int(input("Enter first number: "))

# Take second input from the user
user_input_number_2 = int(input("Enter second number: "))
```

```python
# Compare both user input numbers
if user_input_number_1 == user_input_number_2:
    print("The first number is equal to the second number.")

elif user_input_number_1 > user_input_number_2:
    print("The first number is greater than the second number.")

else:
    print("The second number is greater than the first number.")
```
