## 5.1 Generation calculator Exercise

Create a function named "generation_calculator" to help determine the generation you belong to according to your birth year. Follow the below information:



- Define the function generation calculator. Write conditionals for each generation. For example, if you were born before 1949 you would be a "silent generation".

- Call the function to test it with different year values.
- Print the results

```python
# Define the function
def generation_calculator(year):
    
    if year < 1949:
        return "Silent Generation"
    elif 1949 <= year <= 1968:
        return "Baby Boomer"
    elif 1969 <= year <= 1980:
        return "Generation X"
    elif 1981 <= year <= 1993:
        return "Millennial"
    else:
        return "Generation Z"
    
# Take user input
user_input = int(input("Enter year you were born in: "))

# Print the returned value
print("Your generation is:", generation_calculator(user_input))
```
